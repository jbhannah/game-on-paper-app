```python
import json
import urllib
import pandas as pd
from datetime import datetime

def get_schedule(team, year):
    url = "http://site.api.espn.com/apis/site/v2/sports/football/college-football/teams/{}/schedule?season={}&seasontype=2".format(team,year)
    resp = urllib.request.urlopen(url)
    txt = json.loads(resp.read())['events']
    bowl_url = "http://site.api.espn.com/apis/site/v2/sports/football/college-football/teams/{}/schedule?season={}&seasontype=3".format(team,year)
    bowl_resp = urllib.request.urlopen(bowl_url)
    bowl_txt = json.loads(bowl_resp.read())['events']
    
    return pd.concat([pd.DataFrame(txt), pd.DataFrame(bowl_txt)])

bowl_txt = get_schedule('Florida-State',2016)
bowl_txt
```
